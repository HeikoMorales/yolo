```python
import pandas as pd

# Suponiendo que los datos se cargan desde un archivo CSV
df = pd.read_csv("C:/Users/aritz/OneDrive/Escritorio/MUUUH/Master/Semestre 2/DI/yolo/runs/pose/data/user1_20240429145752.csv")

# Pivotar el DataFrame para obtener las coordenadas en una fila por cada label
df_pivot = df.pivot_table(index=['Session', 'Usuario', 'FrameID'], columns='Label', values=['X', 'Y'], aggfunc='first')

# Renombrar las columnas
df_pivot.columns = ['_'.join(col).strip() for col in df_pivot.columns.values]

# Restablecer el índice
df_pivot.reset_index(inplace=True)

# Mostrar el resultado
df_pivot.head()
```
